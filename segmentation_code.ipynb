# Segmentation Code

The following code segmented the original images into two masks with noise.

We used the following libraries.

```python
from PIL import Image
import cv2
import copy
from matplotlib import pyplot as plt
```

For each image name in our list, we import the name and isolate the identification number (the Smithsonian Institution Digital Asset Management System (DAMS) barcode number assigned to the image) for the image. We stored the identification number in the variables foreground_output and background_output, to use for naming our masks later.

For the sake of this example, we will only perform the segmentation on one image, "example.jpg", which is not associated with a DAMS barcode number.

We imported the image as an array of pixel values in grayscale. We made two copies of this array to turn into our masks.

```python
orig = cv2.imread("example.jpg")
image_name = "example.jpg"
foreground_output = image_name.split(".")[0] + "_foreground.jpg"
background_output = image_name.split(".")[0] + "_background.jpg"
figure_output = image_name.split(".")[0] + "_figure.pdf"
    
orig = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE) #imports the image
    
mask1 = copy.deepcopy(orig) #makes two copies of the array that we can change without changing the original
mask2 = copy.deepcopy(orig)
```

# Mask 1

We used Otsu’s binarization method as implemented in OpenCV to estimate image thresholds. In short, Otsu’s method searches the distribution of pixel values to estimate a threshold that minimizes intra-class variance. We then assigned a binary value (either black, 0 in byte image values, or white, 255) to each pixel according to the Otsu threshold to create the foreground masks.

Anywhere the thresholded image was black, we set the original image stored in mask1 (the copy of the original array) to black (pixel value of 0). Everything else was set to white (pixel value of 255).

```python
ret, thresh = cv2.threshold(orig,0,255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU) 

for i in range(len(thresh)):
    for j in range(len(thresh[0])):
        if(thresh[i,j] == 0): #if the background is black in the segmented copy
            mask1[i,j] = 0 #set the copy of the original image to black at the same location
        if(mask1[i,j] > 0):
            mask1[i,j] = 255

image = Image.fromarray(mask1)
image.save(foreground_output,"JPEG") #saves the first mask
```

# Mask 2

Mask 2 is the inverse of mask 1. Anywhere mask 1 was black we set mask 2 to white and vise versa.

```python
img = cv2.imread(foreground_output, cv2.IMREAD_GRAYSCALE)
#imports the first mask so we can use it to make the second mask

for i in range(len(img)):
    for j in range(len(img[0])):
        if(img[i,j] > 0): #if the first mask is not black at a pixel
            mask2[i,j] = 0 #set that pixel to black in the second mask
        if(mask2[i,j] > 0):
            mask2[i,j] = 255
                
image2 = Image.fromarray(mask2)
image2.save(background_output,"JPEG") #saves the second mask    
```

We saved three images: the foreground mask; the background mask; and a pdf with the original, foreground mask, and background mask side by side for comparison.

```python
#Plotting all three images on their own subplot for side-by-side comparison.

plt.subplot(131) #original image
plt.axis('off')
plt.title("Original Image")
plt.imshow(orig,'gray')

plt.subplot(132) #mask 1 - foreground
plt.axis('off')
plt.title("Mask 1-Foreground")
plt.imshow(mask1,'gray')

plt.subplot(133) #mask 2 - background
plt.axis('off')
plt.title("Mask 2-Background")
plt.imshow(mask2,'gray')

plt.savefig(figure_output)
```
